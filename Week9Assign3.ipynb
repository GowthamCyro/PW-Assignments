Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a powerful Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics. Here are five commonly used plots that you can create using Seaborn, along with their uses:

Scatter Plot (sns.scatterplot()):
Use: Scatter plots are used to visualize the relationship between two numerical variables. They are particularly useful for identifying patterns, clusters, or correlations between data points. Seaborn's scatter plot offers additional customization options, such as adding color hue or size mapping based on a third variable.

Bar Plot (sns.barplot()):
Use: Bar plots are used to compare categorical data. They display rectangular bars with lengths proportional to the values they represent. Seaborn's bar plot can display the mean, median, or other summary statistics of the data. It is useful for comparing multiple categories or showing how a numerical variable varies across different groups.

Histogram (sns.histplot()):
Use: Histograms are used to visualize the distribution of a continuous variable. Seaborn's histogram helps to understand the underlying data distribution, such as whether it is skewed or symmetric, and identify any outliers or gaps in the data. It automatically bins the data and displays the frequency of data points in each bin.

Box Plot (sns.boxplot()):
Use: Box plots provide a visual representation of the distribution of numerical data through quartiles. They are useful for comparing the spread, median, and presence of outliers across different categories or groups. Seaborn's box plot allows you to add color hue or use it in combination with other plots to display complex visualizations.

Heatmap (sns.heatmap()):
Use: Heatmaps are used to display a 2D representation of data, typically as a color-coded matrix. They are helpful in visualizing correlations between variables, pairwise relationships, or the distribution of values across different categories. Seaborn's heatmap supports various customization options, such as annotating cells, changing color maps, and handling missing data.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt

dat = sns.load_dataset("fmri")
```

```python
dat.head()
```

```python
sns.lineplot(x ='timepoint',y ='signal',hue='event',style='region',data=dat)
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

```python
titanic = sns.load_dataset("titanic")

plt.subplot(1,2,1)
sns.boxplot(x='pclass',y='age',data=titanic)
plt.ylabel('Age')

plt.subplot(1,2,2)
sns.boxplot(x='pclass',y='fare',data=titanic)
plt.ylabel('Fare')

plt.tight_layout()
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
diamonds = sns.load_dataset('diamonds')
```

```python
diamonds.head()
```

```python
sns.histplot(data=diamonds,x = 'price',hue='cut', bins=30, kde=True, palette='muted')
plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

```python
iris = sns.load_dataset('iris')
sns.pairplot(data = iris,hue='species')
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
flights_data = sns.load_dataset("flights")

flights_pivot = flights_data.pivot('month','year', "passengers")

plt.figure(figsize=(10, 7))
sns.heatmap(flights_pivot, annot=True, fmt='d', cmap='YlGnBu')

plt.xlabel('Year')
plt.ylabel('Month')
plt.title('Number of Passengers per Month (Flights Dataset)')

plt.show()
```
